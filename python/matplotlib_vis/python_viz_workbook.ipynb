<a name="subplots"></a> 
## Subplots

Purpose: Allows you to place multiple charts in a figure.

matplotlib function: `subplot(nrows, ncols, plot_number)`

`nrows`: The number of rows in the figure.

`ncols`: The number of columns in the figure.

`plot_number`: The placement of the chart (starts at 1).



````python
plt.figure(dpi=130)

# Using the loop variable to set the position of each plot with Titles 

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])
    
# This function helps to keep axis labels and tick values from ploting on top of each other. 

plt.tight_layout()

plt.show()
````

````python
# Create dataset

dist1 = np.random.normal(42,7,1000)
dist2 = np.random.normal(59,3,1000)
````

````python
plt.figure(figsize=(10,4))

# Position 1 is for the histogram 

plt.subplot(1,2,1)
plt.hist(dist1)
plt.title('dist1')
plt.xlabel('Range')
plt.ylabel('Values')

# Postition 2 is for the scatter

plt.subplot(1,2,2)
plt.scatter(dist1, dist2)
plt.title('Scatter plot')
plt.xlabel('dist1')
plt.ylabel('dist2')

plt.tight_layout()

plt.show()
````

````python
years = np.arange(2010,2016)

plt.figure(dpi=130)


# Here new data for the plot is being created with each iteration of the loop. 
# And then it is ploted

for category_num in range(1,17):
    y_vals = np.arange(10,16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals)

# adding limits to make uniform the plotting area
plt.ylim(10,19)
plt.xlim(2010,2015)

# x tick lables 
plt.xticks(years, [str(year) for year in years])

# plot labels 
plt.title('Group growth over time')
plt.xlabel('Years')
plt.ylabel('Counts')

plt.show()
````

````python
years = np.arange(2010,2016)

plt.figure(figsize=(12,12))

# Here the subplot location is using the loop variable to specify where in
# the grid the plot is to go.

for category_num in range(1,17):
    plt.subplot(6,3,category_num)
    y_vals = np.arange(10,16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals)

# limit is used to keep each axis units in sync
    plt.ylim(10,18)
    
    plt.xticks(years, [str(year) for year in years])
    plt.title('Group {}'.format(category_num))
    
plt.tight_layout()

plt.show()
````

# Place legend outside of plot area

## Plot legend under plot area

````python
# Under plot area

years = np.arange(2010,2016)

plt.figure(figsize=(10,10))

for category_num in range(1,17):
    y_vals = np.arange(10,16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals, label = category_num)

plt.ylim(10,19)
plt.xlim(2010,2015)
plt.xticks(years, [str(year) for year in years])

plt.title('Legend under plot', fontsize= 18)
plt.ylabel('Counts', fontsize= 14)
plt.xlabel('Years', fontsize= 14)

# Here we define where outside the ploting area to locate the ledgend
plt.legend(bbox_to_anchor=(.5,-.1), loc = 'upper center', ncol = 8)

plt.savefig('plots/outside_legend01.png', dpi=150, bbox_inches='tight')
````

## Place legend above plot area


````python
# Plot ledgend above plot area

years = np.arange(2010,2016)

plt.figure(figsize=(10,8))

for category_num in range(1,17):
    y_vals = np.arange(10,16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals, label = category_num)

plt.ylim(10,19)
plt.xlim(2010,2015)
plt.xticks(years, [str(year) for year in years])

plt.title('Legend above plot', fontsize= 18)
plt.ylabel('Counts', fontsize= 14)
plt.xlabel('Years', fontsize= 14)

plt.legend(bbox_to_anchor=(.5,1.2), loc = 'upper center', ncol = 8)

plt.savefig('plots/outside_legend02.png', dpi=150, bbox_inches='tight')
````

## Styles

matplotlib functions: `style.available`
`style.context(style)`


````python
# Create three datasets

x1_values = [2012,2013,2014,2015]
y1_values = [4.3,2.5,1.8,4.5]

x2_values = [2012,2013,2014,2015]
y2_values = [2.4,4.4,1.8,2.8]

x3_values = [2012,2013,2014,2015]
y3_values = [2,2,3,5]
            
````

````python
# plot data to look similar to formate used by ggplot2

plt.figure(figsize=(10,8))

# plot data
plt.plot(x1_values,y1_values,
        label='Python',
        lw = 3,
        color = 'red')
plt.plot(x2_values,y2_values,
        label='JavaScript',
        lw = 3)
plt.plot(x3_values,y3_values,
        label = 'R',
        lw = 3)

# plot tick labels
plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
plt.yticks([1,2,3,4,5])

# set common limits
plt.xlim(2012,2015)
plt.ylim(0,6)

# add axis labels
plt.xlabel('Year', fontsize=18)
plt.ylabel('Web Searches', fontsize = 18)

# add legend and plot grind
plt.legend(loc='upper center', ncol = 3, fontsize = 18)
plt.grid(True)

# plt.savefig('plots/web_search_01.png', dpi =150)

plt.show()
````

### Ploting figure with a particular style

````python
plt.figure(figsize=(10,8))

with plt.style.context('ggplot'): 

    plt.plot(x1_values,y1_values,
            label='Python',
            lw = 3,
            color = 'red')
    plt.plot(x2_values,y2_values,
            label='JavaScript',
            lw = 3)
    plt.plot(x3_values,y3_values,
            label = 'R',
            lw = 3)

    plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
    plt.yticks([1,2,3,4,5])

    plt.xlim(2012,2015)
    plt.ylim(0,6)

    plt.xlabel('Year', fontsize=18)
    plt.ylabel('Web Searches', fontsize = 18)

    plt.legend(loc='upper center', ncol = 3, fontsize = 18)
    plt.grid(True)

    plt.savefig('plots/web_search_05.png', dpi =150)
    
    plt.show()
````

### Plot examples of all the styles available as one plot

````python

# create a list of matplotlib styles
style_list = plt.style.available

plt.figure(figsize=(14,25))

# now loop over the list of styles and use subplot to arange output

counter = 1 # initialize counter

for style in style_list:

    with plt.style.context(style): 
 
        plt.subplot(8,4,counter)
        plt.plot(x1_values,y1_values,
                label='Python',
                lw = 2,
                color = 'red')
        plt.plot(x2_values,y2_values,
                label='JavaScript',
                lw = 2)
        plt.plot(x3_values,y3_values,
                label = 'R',
                lw = 2)

        plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
        plt.yticks([1,2,3,4,5])

        plt.xlim(2012,2015)
        plt.ylim(0,6)

        plt.xlabel('Year')
        plt.ylabel('Web Searches')
        plt.title(style)

        plt.legend(loc='upper center', ncol = 3, prop={'size': 6})
        plt.grid(True)
        
        counter += 1

plt.tight_layout()
plt.savefig('plots/styles.png', dpi =150)

plt.show()
````
